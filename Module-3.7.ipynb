**1. Numpy:**

Numpy is a powerful library for numerical operations in Python. It provides support for arrays, matrices, and many mathematical functions.

```python
#Example-1


#Installing Numpy
!pip install numpy

#Import Numpy
import numpy as np

# Creating an array
arr = np.array([1, 2, 3, 4, 5])
print(arr)  # Output: [1 2 3 4 5]

# Creating a 2D array (matrix)
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(matrix)
# Output:
# [[1 2 3]
#  [4 5 6]]

# Array operations
print(arr + 10)  # Output: [11 12 13 14 15]
print(matrix * 2)
# Output:
# [[ 2  4  6]
#  [ 8 10 12]]

# Array slicing
print(arr[1:4])  # Output: [2 3 4]

# Mathematical operations
print(np.sin(arr))  # Output: [0.84147098 0.90929743 0.14112001 -0.7568025  -0.95892427]
print(np.mean(arr)) # Output: 3.0
print(np.sum(matrix)) # Output: 21

```

```python
#Example-2

import numpy as np

# Creating a NumPy array
arr = np.array([1, 2, 3, 4, 5])

# Performing vector operations
arr_squared = arr ** 2

# Linear algebra operations
dot_product = np.dot(arr, arr_squared)

print(dot_product)
```

```python
#Example-3

import numpy as np

# Numerical Arrays
data = np.array([1, 2, 3, 4, 5])

# Vector Operations
squared_data = data ** 2

# Linear Algebra
matrix_A = np.array([[1, 2], [3, 4]])
inverse_A = np.linalg.inv(matrix_A)

# Scientific Computing
signal = np.array([0, 1, 0, -1, 0])
fourier_transform = np.fft.fft(signal)

print(fourier_transform)
```

**2. Pandas**

Pandas is a powerful library for data manipulation and analysis, providing data structures like DataFrames and Series.

 **i. DataFrame**

```python
##Installing Pandas
!pip install pandas

#Import Pandas
import pandas as pd

# Creating a DataFrame from a dictionary
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "Los Angeles", "Chicago"]
}
df = pd.DataFrame(data)
print(df)
# Output:
#       Name  Age         City
# 0    Alice   25     New York
# 1      Bob   30  Los Angeles
# 2  Charlie   35      Chicago

# Accessing columns
print(df["Name"])  # Output: 0    Alice 1      Bob 2  Charlie Name: Name, dtype: object

# Accessing rows
print(df.loc[1])  # Output: Name        Bob Age          30 City    Los Angeles Name: 1, dtype: object

# Adding a new column
df["Country"] = ["USA", "USA", "USA"]
print(df)
# Output:
#       Name  Age         City Country
# 0    Alice   25     New York     USA
# 1      Bob   30  Los Angeles     USA
# 2  Charlie   35      Chicago     USA

# Removing a column
df.drop("Country", axis=1, inplace=True)
print(df)
# Output:
#       Name  Age         City
# 0    Alice   25     New York
# 1      Bob   30  Los Angeles
# 2  Charlie   35      Chicago

```

**ii. Series**

```python
# Creating a Series from a list
s = pd.Series([1, 2, 3, 4, 5])
print(s)
# Output:
# 0    1
# 1    2
# 2    3
# 3    4
# 4    5
# dtype: int64

# Creating a Series from a dictionary
data = {"a": 10, "b": 20, "c": 30}
s = pd.Series(data)
print(s)
# Output:
# a    10
# b    20
# c    30
# dtype: int64

# Accessing elements
print(s["b"])  # Output: 20

# Performing operations on Series
print(s * 2)
# Output:
# a    20
# b    40
# c    60
# dtype: int64
```

```python
#Example-3

import pandas as pd

# Sample data creation
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'City': ['New York', 'Los Angeles', 'New York', 'Los Angeles', 'New York'],
    'Height': [1.65, 1.80, 1.75, 1.70, 1.60],  # Height in meters
    'Weight': [68, 75, 70, 80, 60]  # Weight in kilograms
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculations: Calculate Body Mass Index (BMI)
df['BMI'] = df['Weight'] / (df['Height'] ** 2)

# Grouping and Aggregation: Calculate mean of each numeric column grouped by 'City'
numeric_columns = df.select_dtypes(include='number').columns
city_grouped_mean = df.groupby('City')[numeric_columns].mean()

# Applying Functions: Calculate length of each name
df['Name_Length'] = df['Name'].apply(len)

# Display the results
print("Original DataFrame with BMI and Name Length:")
print(df)
print("\nMean values grouped by City:")
print(city_grouped_mean)
```

**3. Matplotlib and Seaborn**

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Seaborn is a data visualization library based on Matplotlib that provides a high-level interface for drawing attractive statistical graphics.

**i. Matplotlib**

```python
#Installing Matplotlib and Seaborn
!pip install matplotlib

#import Matplotlib
import matplotlib.pyplot as plt

# Simple line plot
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")
plt.show()

# Bar plot
categories = ["A", "B", "C"]
values = [10, 20, 15]

plt.bar(categories, values)
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Bar Plot")
plt.show()
```

```python
#Example-2

import matplotlib.pyplot as plt

# Creating a Line Plot
x = ['A', 'B', 'C', 'D']
y = [25, 40, 30, 35]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()

plt.bar(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar Chart Example")
plt.show()
```

```python
#Example-3

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 8]

fig, ax = plt.subplots()  # Create figure and axes

# Style elements
ax.set_xlabel("X-axis", fontsize=14)
ax.set_ylabel("Y-axis", fontsize=14)
ax.set_title("Customized Line Plot", fontsize=16)

# Line and marker customization
plt.plot(x, y, color='blue', linestyle='-', marker='o', markersize=8)
# Grid and axis limits
ax.grid(True, linestyle='--', linewidth=0.5)
ax.set_xlim(0, 6)
ax.set_ylim(0, 10)

# Legend customization
plt.legend(title="Data", loc='upper left', fontsize=12)
plt.show()
```

**ii. Seaborn**

```python
#Installing Seaborn
!pip install seaborn

#import Seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# Simple heatmap
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Heatmap of Flights Data")
plt.show()

```

```python
#Example-2

import seaborn as sns

# Creating a Scatter Plot
sns.scatterplot(x=x, y=y, hue=y, palette='viridis')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
#Example-3
import seaborn as sns

data = {'Age': [25, 30, 32, 38, 45]}

sns.set_theme(style='darkgrid')  # Set darkgrid theme
sns.distplot(data["Age"], color='skyblue')  # Customize color
plt.xlabel("Age")
plt.ylabel("Density")
plt.title("Age Distribution (Custom Theme)")

plt.show()
```
