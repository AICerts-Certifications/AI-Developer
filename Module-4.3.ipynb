**Exercise 1:**

**Title:** Hands-on: Customer segmentation

**Problem Statement:**
The increasing complexity and volume of consumer data necessitate effective methods for integrating and analyzing customer profiles and transaction histories to enhance business strategies and customer service. The objective of this exercise is to merge and analyze a dataset containing basic customer demographics with their respective purchase histories.

**Steps to be Followed:**

**1.	Data Preparation:**

    a.	Create a DataFrame with customer demographic data.
    b.	Create a DataFrame with customer purchase history.

**2.	Merge DataFrames:**

    a.	Combine the customer demographic data with the purchase history based on a common identifier (CustomerID).

**3.	Data Preprocessing:**

    a.	Handle any missing values.
    b.	Scale features if necessary (e.g., using standardization or normalization).

**4.	Feature Selection:**

    a.	Select relevant features for clustering, such as 'Age' and 'Amount'.

**5.	Apply K-means Clustering:**

    a.	Choose the number of clusters (K).
    b.	Apply the K-means clustering algorithm on the selected features.

**6.	Assign Cluster Labels:**

    a.	Assign the resulting cluster labels to each customer.

**7.	Visualize the Clusters:**

    a.	Use a scatter plot to visualize the clusters based on the selected features.



```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 1: Data Preparation
# Create a DataFrame with customer data
customer_data = pd.DataFrame({
    'CustomerID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [30, 25, 35, 28, 40],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female'],
    'Location': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami']
})

# Create a DataFrame with purchase history
purchase_history = pd.DataFrame({
    'CustomerID': [1, 2, 3, 3, 4, 5, 1, 2, 3, 4],
    'PurchaseID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Product': ['Laptop', 'Smartphone', 'Tablet', 'Smartwatch', 'Headphones',
                'Camera', 'Smartphone', 'Tablet', 'Smartwatch', 'Headphones'],
    'Amount': [1200, 800, 500, 300, 200, 1000, 750, 450, 350, 180]
})

# Step 2: Merge DataFrames
# Merge customer data with purchase history
merged_data = pd.merge(customer_data, purchase_history, on='CustomerID')

print(merged_data)

# Step 3: Data Preprocessing
# Handle any missing values (if present) - not needed in this case as there are no missing values

# Step 4: Feature Selection
# Features for clustering (e.g., Age and Amount)
features_for_clustering = ['Age', 'Amount']
data_for_clustering = merged_data[features_for_clustering]

# Step 5: Apply K-means Clustering
# Choose the number of clusters (K)
k = 3
kmeans = KMeans(n_clusters=k)
kmeans.fit(data_for_clustering)

# Step 6: Assign Cluster Labels
# Assign cluster labels to each customer
merged_data['Cluster'] = kmeans.labels_

# Step 7: Visualize the Clusters
plt.scatter(merged_data[features_for_clustering[0]], merged_data[features_for_clustering[1]], c=merged_data['Cluster'], cmap='viridis')
plt.xlabel(features_for_clustering[0])
plt.ylabel(features_for_clustering[1])
plt.title('Customer Segmentation')
plt.show()
```

**Code Analysis and Explanation:**

**1.	Data Preparation:**

    a.	Two DataFrames are created: customer_data with demographic information and purchase_history with transaction details.

**2.	Merge DataFrames:**
    
    a.	The pd.merge function merges the two DataFrames on the CustomerID column to create a consolidated dataset (merged_data).

**3.	Data Preprocessing:**
    
    a.	This step includes handling missing values and scaling features if needed. In this script, it is assumed that there are no missing values.

**4.	Feature Selection:**
    
    a.	Relevant features ('Age' and 'Amount') are selected for clustering and stored in data_for_clustering.

**5.	Apply K-means Clustering:**
    
    a.	The K-means clustering algorithm is applied to the selected features. The number of clusters is set to 3 (k = 3).

**6.	Assign Cluster Labels:**
   
    a.	Cluster labels assigned by the K-means algorithm are added to the merged_data DataFrame.

**7.	Visualize the Clusters:**
    
    a.	A scatter plot is generated to visualize the clusters based on 'Age' and 'Amount', with different colors representing different clusters.


**Exercise 2:**

**Title:** Hands-on: Market Research

**Problem Statement:**

In the competitive landscape of retail marketing, understanding customer demographics and purchasing behaviors is crucial for optimizing marketing strategies and improving customer satisfaction. The challenge lies in effectively segmenting the customer base into distinct groups that share similar characteristics without prior labeling of data, which is often the case in real-world datasets. The objective of this project is to employ an unsupervised learning approach, specifically K-means clustering, to analyze and segment a simulated dataset based on customer features such as age and annual spending.

**Steps to be Followed:**
**1.	Generate Synthetic Data:**
    
    a.	Create synthetic data representing two features: Age and Annual Spending Score.

**2.	Prepare Data for Clustering:**
    
    a.	Combine the generated features into a single dataset suitable for clustering.

**3.	Apply K-means Clustering:**

    a.	Choose the number of clusters (K).
    b.	Fit the K-means clustering algorithm to the data.
    c.	Predict cluster labels for the data points.

**4.	Visualize the Clusters**

    a.	Use a scatter plot to visualize the clusters, with different colors representing different clusters.
    b.	Highlight the cluster centers.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Generate Synthetic Data
# Assume these are two features: Age and Annual Spending
np.random.seed(42)  # For reproducibility
age = np.random.normal(40, 10, 200)  # Normally distributed ages with mean 40 and std dev 10
spending_score = np.random.normal(50, 20, 200)  # Spending score from 0 to 100

# Step 2: Prepare Data for Clustering
data = np.column_stack((age, spending_score))

# Step 3: Apply K-means Clustering
# Choose the number of clusters (K)
kmeans = KMeans(n_clusters=4, random_state=42)  # 4 clusters
kmeans.fit(data)  # Fit the K-means algorithm to the data
y_kmeans = kmeans.predict(data)  # Predict cluster labels for data points

# Step 4: Visualize the Clusters
plt.figure(figsize=(10, 6))
plt.scatter(data[:, 0], data[:, 1], c=y_kmeans, cmap='viridis', marker='o')  # Plot data points with cluster colors
centers = kmeans.cluster_centers_  # Get the coordinates of cluster centers
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')  # Plot cluster centers
plt.title('Customer Segments')
plt.xlabel('Age')
plt.ylabel('Annual Spending Score')
plt.colorbar(label='Cluster')
plt.show()
```

**Code Analysis and Explanation:**

**1.	Generate Synthetic Data:**

    a.	Synthetic data for customer age and annual spending score is generated using normal distributions. The np.random.normal function creates 200 samples for each feature with specified mean and standard deviation values.

**2.	Prepare Data for Clustering:**

    a.	The generated features (age and spending score) are combined into a single dataset using np.column_stack.

**3.	Apply K-means Clustering:**

    a.	The number of clusters (K) is set to 4. The KMeans algorithm is initialized with n_clusters=4 and a random state for reproducibility.
    b.	The K-means algorithm is fitted to the data using kmeans.fit(data).
    c.	Cluster labels for each data point are predicted using kmeans.predict(data).

**4.	Visualize the Clusters:**

    a.	A scatter plot is created to visualize the clusters. Data points are colored according to their cluster labels.
    b.	Cluster centers are highlighted using red markers.
    c.	The plot is labeled with titles and axes labels, and a color bar is added to indicate the cluster labels.


**Exercise 3:**

**Title:** Hands-on: Anomaly Detection

**Problem Statement:**

In financial transactions, anomaly detection is crucial for identifying potentially fraudulent activities and ensuring the security of financial systems. The goal of this project is to develop an unsupervised learning model that can effectively identify unusual patterns or anomalies in transactional data, which often do not conform to expected behaviors. This kind of detection is vital for preventing fraud, which can result in significant financial losses and erode customer trust.

**Steps to be Followed:**

**1.	Generate Synthetic Data:**
    
    a.	Create synthetic data representing normal and anomalous transaction amounts.

**2.	Prepare Data for Modeling:**
    
    a.	Reshape the transaction data to fit the model.

**3.	Initialize and Fit the Isolation Forest:**
    
    a.	Configure the Isolation Forest algorithm with appropriate parameters.
    b.	Fit the model to the transaction data.
    c.	Predict anomalies in the transaction data.

**4.	Visualize the Results:**
    
    a.	Create a scatter plot to visualize the transaction amounts.
    b.	Color-code normal and anomalous transactions for easy identification.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Step 1: Generate Synthetic Data
# Seed for reproducibility
np.random.seed(42)

# Generate synthetic transaction data
# Normal transactions (mean=100, std=10, count=500)
# Anomalous transactions (mean=200, std=20, count=50)
transaction_amounts = np.concatenate([np.random.normal(100, 10, 500), np.random.normal(200, 20, 50)])

# Step 2: Prepare Data for Modeling
# Reshape data for modeling
transaction_amounts = transaction_amounts.reshape(-1, 1)

# Step 3: Initialize and Fit the Isolation Forest
# Initialize Isolation Forest with 100 estimators and contamination of 0.1
clf = IsolationForest(n_estimators=100, contamination=0.1, random_state=42)
# Fit the model and predict anomalies
predictions = clf.fit_predict(transaction_amounts)

# Step 4: Visualize the Results
# Plotting the results
plt.figure(figsize=(12, 6))
plt.scatter(np.arange(len(transaction_amounts)), transaction_amounts, c=predictions, cmap='coolwarm', label='Normal = Blue, Anomaly = Red')
plt.xlabel('Transaction Index')
plt.ylabel('Transaction Amount')
plt.title('Transaction Amount Anomaly Detection')
plt.legend(["Normal", "Anomaly"])
plt.show()
```

**Code Analysis and Explanation:**

**1.	Generate Synthetic Data:**

    a.	A seed is set for reproducibility using np.random.seed(42).
    b.	Normal transactions are generated with a mean of 100 and standard deviation of 10 for 500 samples.
    c.	Anomalous transactions are generated with a mean of 200 and standard deviation of 20 for 50 samples.
    d.	Both normal and anomalous transactions are concatenated into a single dataset.

**2.	Prepare Data for Modeling:**

    a.	The transaction amounts are reshaped into a 2D array using reshape(-1, 1) to fit the model's expected input format.

**3.	Initialize and Fit the Isolation Forest:**

    a.	The Isolation Forest is initialized with 100 estimators and a contamination rate of 0.1, which specifies the proportion of anomalies in the data.
    b.	The model is fitted to the transaction data and predictions are made using fit_predict.

**4.	Visualize the Results:**

    a.	A scatter plot is created to visualize the transaction amounts.
    b.	Data points are color-coded based on the predictions, with normal transactions in one color and anomalies in another.
    c.	The plot is labeled with titles and axis labels, and a legend is added to differentiate between normal and anomalous transactions.

**Title:** Visualizing high-dimensional data for easy analysis

**Problem Statement:**
In today's data-driven world, the complexity and dimensionality of data have increased dramatically, presenting challenges in data analysis and visualization. High-dimensional data, which are typical in domains such as genomics, finance, and digital marketing, can be difficult to interpret and explore using traditional two-dimensional visualization techniques. The objective of this exercise is to employ dimensionality reduction techniques to transform high-dimensional data into lower-dimensional spaces that can be easily visualized and analyzed. This process will help uncover hidden patterns, trends, and relationships which are not apparent in the original high-dimensional space, facilitating better decision-making and insights.

**Steps to be Followed:**

**1.	Generate Synthetic High-Dimensional Data:**

    a. Create a dataset with 100 samples and 10 features, with values drawn from a normal distribution.

**2.	Standardize the Features:**

    a. Standardize the dataset to have a mean of 0 and a standard deviation of 1 for each feature.

**3.	Apply PCA for Dimensionality Reduction:**

    a. Use PCA to reduce the dimensionality of the data from 10 dimensions to 2 principal components.

**4.	Visualize the Principal Components:**

    a. Create a scatter plot to visualize the data in the reduced 2-dimensional space.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Step 1: Generate Synthetic High-Dimensional Data
# Seed for reproducibility
np.random.seed(42)

# Generate synthetic high-dimensional data (100 samples, 10 features)
data = np.random.normal(0, 1, (100, 10))

# Step 2: Standardize the Features
# Standardizing the features
data = StandardScaler().fit_transform(data)

# Step 3: Apply PCA for Dimensionality Reduction
# PCA for dimensionality reduction to 2 components
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data)

# Step 4: Visualize the Principal Components
# Visualization of the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(principal_components[:, 0], principal_components[:, 1], alpha=0.7, edgecolors='w', s=80)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of High-Dimensional Data')
plt.grid(True)
plt.show()
```

**Code Analysis and Explanation:**

**1.	Generate Synthetic High-Dimensional Data:**

    a.	A random seed is set for reproducibility using np.random.seed(42).
    b.	Synthetic data is generated using np.random.normal(0, 1, (100, 10)), which creates 100 samples with 10 features each, with values drawn from a normal distribution with mean 0 and standard deviation 1.

**2.	Standardize the Features:**

    a.	The StandardScaler is used to standardize the features so that each feature has a mean of 0 and a standard deviation of 1. This is important for PCA as it ensures that all features contribute equally to the result.

**3.	Apply PCA for Dimensionality Reduction:**

    a.	PCA is applied to the standardized data to reduce its dimensionality from 10 dimensions to 2 principal components. The fit_transform method computes the principal components and transforms the data.

**4.	Visualize the Principal Components:**

    a.	A scatter plot is created to visualize the data in the 2-dimensional space defined by the first two principal components.
    b.	Points are plotted with a slight transparency (alpha=0.7) and white edges (edgecolors='w') for better visualization.
    c.	The plot is labeled with titles and axis labels, and a grid is added for better readability.
